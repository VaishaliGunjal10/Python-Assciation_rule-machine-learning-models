```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
grocery=[]
with open ("groceries.csv") as f:
    grocery=f.read()
grocery=grocery.split('\n')
grocery
```

```python
grocery_list=[]
for i in grocery:
    grocery_list.append(i.split(','))
grocery_list
```

```python
all_grocery_list=[i for item in grocery_list for i in item]
all_grocery_list
 
     
```

```python
from collections import Counter,OrderedDict
item_counter=Counter(all_grocery_list)
item_counter
item_counter=sorted(item_counter.items(), key=lambda y:y[1])
item_counter
```

```python
counter=list(reversed([i[1] for i in item_counter]))
counter
```

```python
counter_i=list(reversed([i[0] for i in item_counter]))
counter_i
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
plt.bar(x=range(0,5),height=counter[0:5],color='rgbkymc')
plt.xlabel("Items")
plt.ylabel("counter")
plt.xticks(list(range(0,5),),counter_i [0:5])
```

```python
series=pd.DataFrame(pd.Series(grocery_list))
series = series.iloc[:9835,:]
series.columns=["Transactions"]
series
```

```python
X =  series['Transactions'].str.join(sep='*').str.get_dummies(sep='*')
X
```

```python
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending=False,inplace=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules
 
```
