```python
from mlxtend.frequent_patterns import apriori,association_rules
import pandas as pd
```

```python
movies=[]
with open ("my_movies.csv") as f:
    movies=f.read()
movies=movies.split('\n')
movies
movies_list=[]
for i in movies:
     
    movies_list.append(i.split(','))
movies_list

 
 
    
```

```python
all_movies_list1 =[i for item in movies_list  for i in item]
all_movies_list1 
```

```python
from collections import Counter,OrderedDict
item_counter=Counter(all_movies_list1 )
item_counter
item_counter=sorted(item_counter.items(), key=lambda y:y[1])
item_counter
```

```python
counter=list(reversed([i[1] for i in item_counter]))
counter
```

```python
counter_i=list(reversed([i[0] for i in item_counter]))
counter_i
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
plt.bar(x=range(0,5),height= counter[0:5],color='rgbkymc')
plt.xlabel("Items")
plt.ylabel("counter")
plt.xticks(list(range(0,5),), counter_i  [0:5])
```

```python
series=pd.DataFrame(pd.Series(movies_list ))
series=series.iloc[:11,:]
series.columns=["Transactions"]
series
```

```python
X =  series['Transactions'].str.join(sep='*').str.get_dummies(sep='*')
X
```

```python
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending=False,inplace=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules
 
```

```python
import random
import matplotlib.pyplot as plt


for i in range (len(rules['support'])):
    rules['support'][i] = rules['support'][i] + 0.0025 * (random.randint(1,10) - 5) 
    rules['confidence'][i] = rules['confidence'][i] + 0.0025 * (random.randint(1,10) - 5)

plt.scatter(rules['support'], rules['confidence'],   alpha=0.5, marker="*")
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
